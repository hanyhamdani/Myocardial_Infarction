# Basic Classification Model

```python
y = leth_data['LET_IS'].values
X = (leth_data.drop(columns=['LET_IS'])).values
#X= top_ten_feats.values
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y, test_size =0.3, random_state= 100)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, recall_score,precision_score
```

```python
def run_model(X_Value):
    y = leth_data['LET_IS'].values
    X= X_Value.values
    
    xtrain, xtest, ytrain, ytest = train_test_split(X,y, test_size =0.3, random_state= 100)
    
    classifiers =[]

    svc = SVC(probability=True)
    classifiers.append(svc)

    rf = RandomForestClassifier(n_estimators=500)
    classifiers.append(rf)

    knc = KNeighborsClassifier(n_neighbors=5)
    classifiers.append(knc)

    dt = DecisionTreeClassifier(max_features='auto')
    classifiers.append(dt)

    gbc = GradientBoostingClassifier(max_features='auto')
    classifiers.append(gbc)

    for clf in classifiers:
        model = clf.fit(xtrain,ytrain)
        mpred = model.predict(xtest)
        print(clf)
        print("Overall Accuracy:",accuracy_score(ytest, mpred))
        print("Overall Precision:",precision_score(ytest, mpred, average='macro'))
        print("Overall Recall:",recall_score(ytest, mpred, average='macro'))
        print("\n")
```

The best performance is achieved by Random Forest on this data and has accuracy score of 89%, while recall only 22.6%
overall best is by Gradient Boosting with accuracy ~87% and recall 25%
